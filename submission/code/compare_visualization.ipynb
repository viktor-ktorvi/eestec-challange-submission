## Between class comparison

```python
raw_quasi_static_files = get_filenames(data_folder, subfolder='quasi-static', extension=extension, max_files=1, shuffle=True)
quasi_static_files = [os.path.join('quasi-static', f) for f in raw_quasi_static_files]
print(quasi_static_files)

raw_moving_files = get_filenames(data_folder, subfolder='moving', extension=extension, max_files=1, shuffle=True)
moving_files = [os.path.join('moving', f) for f in raw_moving_files]
print(moving_files)
```
